```python
# Import all necessary independences
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# Drop a column that is not useful
df.drop(columns=['Patient_ID','ER status','PR status','Date_of_Surgery','Date_of_Last_Visit'], inplace=True)
```

```python

# Handle missing values for categorical columns
cat_cols = ['Gender','Tumour_Stage','Histology','HER2 status',
            'Surgery_type', 'Patient_Status']
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Handle missing values for numerical columns
num_cols = ['Age','Protein1','Protein2','Protein3','Protein4']
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
#dropping the duplicates row
df.drop_duplicates(inplace=True)
```

```python
#Handle categorical variables
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label_count = 0

for col in df.columns:
    if df[col].dtype == 'object':


        label.fit(df[col])
        df[col] = label.transform(df[col])


        label_count += 1
        print('%d columns were label encoded.' % label_count)
```

```python
df.head()
```

```python
#splitting the data
X = df.drop(["Patient_Status"], axis=1).values
y = df['Patient_Status'].values
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 20)
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
 #Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
'Accuracy:', accuracy_score(y_test, y_pred)
```

```python
print('Training set score: {:.4f}'.format(model.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(model.score(X_test, y_test)))
```
